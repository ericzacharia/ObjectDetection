```python
import matplotlib.pyplot as plt
import os
import cv2

for directory in ['train', 'valid']:
    dataset_dir = f'dataset/{directory}/images'

    # Get the list of image files in the dataset directory
    image_files = [file for file in os.listdir(dataset_dir) if file.endswith('.jpg') or file.endswith('.png')]

    # Create empty lists to store the image sizes
    image_sizes = []

    # Iterate over each image file and append its size to the list
    for image_file in image_files:
        image_path = os.path.join(dataset_dir, image_file)
        image = cv2.imread(image_path)
        try:
            image_size = image.shape[:2]  # Get the height and width of the image
            image_sizes.append(image_size)
        except AttributeError:
            print(f'image {image_file} has no shape')

    # Extract the width and height values from the image sizes
    # widths = [size[1] for size in image_sizes]
    # heights = [size[0] for size in image_sizes]
    aspect_ratios = [size[1] / size[0] for size in image_sizes]

    # create a histogram of the aspect ratios
    plt.figure()
    plt.hist(aspect_ratios)
    plt.xlabel('Aspect Ratio')
    plt.ylabel('Frequency')
    plt.title('Aspect Ratio Distribution')
    plt.show()
```
