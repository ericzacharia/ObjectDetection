```python
import os

def process_images(input_directory, output_directory):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for filename in os.listdir(input_directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Check for image files
            label_file_path = os.path.join(output_directory, os.path.splitext(filename)[0] + ".txt")
            with open(label_file_path, 'w') as f:
                # Write normalized coordinates for a bounding box that covers the entire image
                f.write("0 0.5 0.5 1.0 1.0")

# Directories
input_directory = "LPs"       # Replace with the path to your 'LPs' directory
output_directory = "LPs_labels"  # Replace with the path to your 'LPs_labels' directory

process_images(input_directory, output_directory)
```

```python
import os
import random
import shutil

images_dir = 'LPs'
labels_dir = 'LPs_labels'
images = sorted(os.listdir(images_dir))
labels = sorted(os.listdir(labels_dir))
len(images), len(labels)
count = 0
for image, label in zip(images, labels):
    count += 1
    if count > len(os.listdir('dataset/train/images'))+len(os.listdir('dataset/valid/images')):
        break
    # check if it is a file:
    if not os.path.isfile(os.path.join(images_dir, image)):
        continue
    if random.random() <= 0.8:
        shutil.move(os.path.join(images_dir, image), 'dataset/train/images')
        shutil.move(os.path.join(labels_dir, label), 'dataset/train/labels')
    else:
        shutil.move(os.path.join(images_dir, image), 'dataset/valid/images')
        shutil.move(os.path.join(labels_dir, label), 'dataset/valid/labels')

```
